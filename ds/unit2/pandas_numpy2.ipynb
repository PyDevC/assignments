```python
df = pd.read_csv('u_user.csv', sep='|')
df.head(25)
```

```python
df.occupation.unique()
```

```python
# mean and standard deviation of age columns
mean = df.age.mean()
std = df.age.std()
print(f"Mean: {mean}\nStandard deviation: {std}")
```

```python
# replace 34 by nan
# in age column

df['age'].replace(to_replace=35, value=np.nan, inplace=True)
df.isnull().sum()
```

```python
# sort values by age
# ascending
df.sort_values(by='age', inplace=True)
df.head()
```

```python
# descending
df.sort_values(by='age', ascending=False, inplace=True)
df.head()
```

```python
df.query('age > 20 and age < 50')
```

```python
# new column with square of age
df['age_square'] = df.age.pow(2)
```
