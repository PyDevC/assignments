# Numpy & Pandas

```python
import numpy as np
import pandas as pd
```

```python
# evenly spaced array
arr = np.linspace(0.0,10.0, num=20)
arr

# Reason for uneven numbers is because 10 is inclusive. maybe
```

```python
data = pd.DataFrame(np.random.randint(0, 100, (100,5)))
data.head()
```

```python
# loading data to csv format
data.to_csv('randomdata.csv')
```

```python
# mean of each column
data.agg("mean")
```

```python
# filter rows from each column wich has value greater than 50
# of column 0
data[data[0] > 50]
```

```python
# concatenate dataframe horizontally
df1 = pd.DataFrame(np.random.randint(0, 100, (100,5)))
df2 = pd.DataFrame(np.random.randint(0, 100, (100,10)))

df = pd.concat([df1, df2], axis=1)
df.head()
```

```python
# add new column sum which has sum of all the columns
df['sum'] = df.sum(axis=1)
```
